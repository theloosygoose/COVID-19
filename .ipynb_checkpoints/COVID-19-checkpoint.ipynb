```python
tt = """Ala / A 	GCU, GCC, GCA, GCG
Ile / I 	AUU, AUC, AUA
Arg / R 	CGU, CGC, CGA, CGG; AGA, AGG
Leu / L 	CUU, CUC, CUA, CUG; UUA, UUG 
Asn / N 	AAU, AAC
Lys / K 	AAA, AAG
Asp / D 	GAU, GAC
Met / M 	AUG
Phe / F 	UUU, UUC
Cys / C 	UGU, UGC
Pro / P 	CCU, CCC, CCA, CCG
Gln / Q 	CAA, CAG
Ser / S 	UCU, UCC, UCA, UCG; AGU, AGC
Glu / E 	GAA, GAG
Thr / T 	ACU, ACC, ACA, ACG
Trp / W 	UGG
Gly / G 	GGU, GGC, GGA, GGG
Tyr / Y 	UAU, UAC
His / H 	CAU, CAC 
Val / V 	GUU, GUC, GUA, GUG
STOP 	UAA, UGA, UAG
""".strip()
dec = {}
for t in tt.split('\n'):
    k,v = t.split('\x09')
    if '/' in k:
        k = k.split('/')[-1].strip()
    v = v.replace(",", "").replace(";","").lower().replace("u", "t").split(" ")
    for vv in v:
        dec[vv] = k
dec
```

Genome of COVID-19

```python
for a in '\n':
    cc = cc.replace(a, '')
cc
```

Parsed Amino Acids from genome

```python
# Parsed Amino acids from genome

aa = []
for rf in range(3): 
    for i in range(rf, len(cc)-3,3):
        aa.append(dec[cc[i:i+3]])
aa = ''.join(aa)
aa = aa.split('STOP')
```

Filtering out 1 amino acid sequences < 100

```python
bb = list(filter(lambda x: len(x) > 50, aa))

bb
```
